```python
same_amount_rows = raw_data.dropna()
same_amount_rows.info()
```

```python
dummies = pandas.get_dummies(same_amount_rows["ocean_proximity"])
dummies
```

```python
dummies = pandas.get_dummies(same_amount_rows["ocean_proximity"], dtype='int')
dummies
```

```python
adding_dummies = same_amount_rows.join(dummies)
adding_dummies
```

```python
clean_data = adding_dummies.drop('ocean_proximity', axis=1)
clean_data.info()
```

```python
clean_data.corr()
```

```python
seaborn.set_theme(rc={'figure.figsize': (15,8)})
seaborn.heatmap(clean_data.corr(), annot=True, cmap='YlGnBu')
```
