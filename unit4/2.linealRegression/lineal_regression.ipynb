```python
x = clean_data.drop(["median_house_value"], axis=1)
y = clean_data["median_house_value"]
```

```python
# Divide the data between two parts: 
# 1. training set
# 2. testing set
from sklearn.model_selection import train_test_split
x_training, x_testing, y_training, y_testing = train_test_split(x, y, test_size=.2) #.2: 20%
```

```python
# x has 14 columns, and y only one
print(x_training.shape)
print(y_training.shape)
print(x_testing.shape)
print(y_testing.shape)
```

```python
# Creating a new model with linear regression
from sklearn.linear_model import LinearRegression
my_model = LinearRegression()
```

```python
# Training the previous model
my_model.fit(x_training, y_training)
```

```python
# Testing the model
predictions = my_model.predict(x_testing)
predictions
```

```python
# Convert the last var in a table
pandas.DataFrame(predictions)
```

```python
# compare training values against testing values
comparisons = {"Predictions Values": predictions, "Real Values": y_testing}
dt = pandas.DataFrame(comparisons)
dt
```

```python
# Overfitting -> Sobreajuste:
# It's when the model has great results with the training data but bad results with the testing data
print(my_model.score(x_training, y_training))
print(my_model.score(x_testing, y_testing))
# Note: the distance each other is not far... so that the model doesn't have Overfitting
```

```python
# Get the error average between our predictions and the real values
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_testing, predictions)
mse
# We use pow the data to 2 (**2) for two reasons:
# 1. the negative numbers change to positive
# 2. the difference be larger
# Altough we can reduce the number with square root
```

```python
import numpy
rmse = numpy.sqrt(mse)
rmse
# This is the average difference between our predictions and real values
```
