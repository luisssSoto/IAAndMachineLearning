```python
#reading file csv with pandas
import pandas
data = pandas.read_csv('./heightAndWeight.csv')
```

```python
#showing the whole information
data.info()
```

```python
#showing the headers and some rows
data.head()
```

```python
#Show the data in Scatter Plot graph
import seaborn
seaborn.scatterplot(x = 'weight', y = 'height', data = data, hue = 'weight')
```

```python
#Separate the dependient variables and independent variables
independent_variable = data['height']
dependent_variable = data['weight']
print(f'Dependent Variable: {dependent_variable}')
print(f'Independent Variable: {independent_variable}')
```

```python
# Reshaping the values of the variables
characteristic = independent_variable.values.reshape(-1, 1)
print(f'Characteristic height: {characteristic}')
tag = dependent_variable.values.reshape(-1, 1)
print(f'Tag weight: {tag}')
```

```python
#Creating a new model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
#Training the model
model.fit(characteristic, tag)
```

```python
# Testing the model
heights = [201, 130, 189, 177, 200, 198]
predictions = []
for height in heights:
    predictions.append(model.predict([[height]]))
print(predictions)
```

```python
#Score the exactitud of the model
model.score(characteristic, tag)
```
