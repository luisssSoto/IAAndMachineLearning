```python
#Reading data with pandas
import pandas
data = pandas.read_csv('./celsiusToFahrenheit.csv')
```

```python
data.info()
```

```python
#show some data and headers
data.head()
```

```python
import seaborn

seaborn.scatterplot(x = 'celsius', y = 'fahrenheit', data = data, hue = 'fahrenheit', palette = 'coolwarm' )
```

```python
#Separate the data
X = data['celsius']
y = data['fahrenheit']
print(f'Celsius: {X}')
print(f'Fahrenheit: {y}')
```

```python
#Reshape the data
X.values
X.values.reshape(-1, 1)
#Assigning the data to variables
characteristics = X.values.reshape(-1, 1)
print(f'Characteristics: {characteristics}')
tags = y.values.reshape(-1, 1)
print(f'Tags: {tags}')
```

```python
#Creating a new model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# Training the model
model.fit(characteristics, tags)
```

```python
# Doing predictions
celsius = 8
prediction = model.predict([[celsius]])
```

```python
# Knowing how exact our model predicts
model.score(characteristics, tags)
'''In real proyects never will find 1.0 because always will be a certain degree of error,
so 1.0 means that the model was perfectly trained, so the error degree is almost null'''
```
